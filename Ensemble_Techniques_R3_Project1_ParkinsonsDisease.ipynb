```python
parkinsons_df['name'] = parkinsons_df['name'].apply(pd.to_numeric, errors='coerce')
parkinsons_df.dtypes
parkinsons_df.drop('name',axis =1,inplace = True)
```

```python
# Now build the model
#Q4 Spliting the data set into test and train (70:30)

array = parkinsons_df.values
X = array[:,0:21] # select all rows and first 23 columns which are the attributes
Y = array[:,22]   # select all rows and the 24th column which is the classification "Yes", "No" for parkinsons disease
test_size = 0.30 # taking 70:30 training and test set
seed = 7  # Random numbmer seeding for reapeatability of the code
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)
```

```python
dt_model = DecisionTreeClassifier(criterion = 'entropy' )
dt_model.fit(X_train, Y_train)
```

```python
print(dt_model.score(X_test , Y_test))
print(dt_model.score(X_train , Y_train))
# Since the training data is 100% accurate and not the test data, this is a clear case of Overfit model

from sklearn import metrics
Y_predict = dt_model.predict(X_test)
print(metrics.confusion_matrix(Y_test, Y_predict))

Cross tab is not working
# crosstab
pd.crosstab(parkinsons_df.status, parkinsons_df.MDVP:Jitter(%),margins=True)
```
